```python
#standardize of Data & eliminating string  type Cloumn 
df1=df.iloc[:,2:]
name=df1.columns
nm=df1[name]
scaler=StandardScaler()
scaler.fit(nm)
```

```python
scaledata=scaler.transform(nm)
scaledata.shape
```

```python
df1.describe()
```

```python
pca=PCA()
pca.fit(scaledata)
plt.figure(figsize=(5,5))
f=range(pca.n_components_)
#print(f)
print(pca.explained_variance_)
plt.bar(f,pca.explained_variance_)
plt.xlabel('PCA Features')
plt.ylabel('Variance')
plt.title('Principal Component Analysis')
```

```python
pca=PCA(n_components=1)
pca.fit(scaledata)
```

```python
x_pca=pca.transform(scaledata)
```

```python
x_pca.shape
x_pca
pcaDf = pd.DataFrame(data = x_pca, columns = ['pca1'])
pcaDf
```

```python
#convert to datetime datatype
df['Date'] = pd.to_datetime(df['Date'])
anomaly['Date'] = pd.to_datetime(anomaly['Date'])
```

```python
#Quantile and IQR Calculation
q1_pca1, q3_pca1 = pcaDf['pca1'].quantile([0.25, 0.75])
iqr_pca1 = q3_pca1 - q1_pca1
#Outlier Boundaries :Calculate upper and lower bounds for outlier for pca1
lower_pca1 = q1_pca1 - (1.5 * iqr_pca1)
x_lower_pca1 = q1_pca1 - (3 * iqr_pca1)#extreme lower bound
upper_pca1 = q3_pca1 + (1.5 * iqr_pca1)
x_upper_pca1 = q3_pca1 + (3 * iqr_pca1)#extreme upper bound
#Anomaly Detection: Filter out the outliers from the pca1
df_outlier['anomaly_pca1'] = ((pcaDf['pca1'] > upper_pca1) | (pcaDf['pca1'] < lower_pca1)).astype('int')
#Extracting Anomalies
anomaly = df[df_outlier['anomaly_pca1'] == 1]
#Plotting Anomalies:plotting outlier/anomaly with original data
plt.figure(figsize=(18,6))
#for date plotiing
plt.plot_date(df['Date'], df['Open'], color='blue', label='Normal')
plt.plot_date(anomaly['Date'], anomaly['Open'], linestyle='none', marker='.', color='red', label='Anomaly')
plt.xlabel('Date')
plt.ylabel('Open Reading')
plt.title('Open Anomalies')
plt.legend(loc=4)
plt.show();
```

```python
#Taking extreme upper and lower bound. Therefore taking extreme outliers  for column Open
df_outlier['anomaly_pca1'] = ((pcaDf['pca1']>x_upper_pca1) | (pcaDf['pca1']<x_lower_pca1)).astype('int')
anomaly = df[df_outlier['anomaly_pca1'] == 1]
plt.figure(figsize=(18,6))
plt.plot_date(df['Date'], df['Open'], color='blue', label='Normal')
plt.plot_date(anomaly['Date'],anomaly['Open'], linestyle='none', marker='.', color='red',label='Anomaly')
plt.xlabel('Date and Time')
plt.ylabel('Open Reading')
plt.title('Open Anomalies')
plt.legend(loc=4)
plt.show();
```
